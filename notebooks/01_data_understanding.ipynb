# Dataset Understanding Summary

## Delays Dataset

## Airports Dataset
- Contains airport metadata

```python
import pandas as pd
import numpy as np

flights = pd.read_csv('../data/raw/flights.csv')
delays = pd.read_csv('../data/raw/delay.csv')
airports = pd.read_csv('../data/raw/airports.csv')


flights['FL_DATE'] = pd.to_datetime(flights['FL_DATE'])

flights['YEAR'] = flights['FL_DATE'].dt.year
flights['MONTH'] = flights['FL_DATE'].dt.month
flights['DAY'] = flights['FL_DATE'].dt.day
flights['DAY_NAME'] = flights['FL_DATE'].dt.day_name()

delay_cols = [
    'DELAY_DUE_CARRIER',
    'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS',
    'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]

flights[delay_cols] = flights[delay_cols].fillna(0)

flights['CANCELLATION_CODE'] = flights['CANCELLATION_CODE'].fillna('Not Cancelled')

flights.duplicated().sum()

flights['CANCELLED'].value_counts()

flights.to_csv(
    '../data/cleaned/flights_cleaned.csv',
    index=False
)
```

```python
active_flights = flights[flights['CANCELLED'] == 0]
cancelled_flights = flights[flights['CANCELLED'] == 1]
```

```python
active_flights.to_csv('../data/processed/active_flights.csv', index=False)
cancelled_flights.to_csv('../data/processed/cancelled_flights.csv', index=False)
```
