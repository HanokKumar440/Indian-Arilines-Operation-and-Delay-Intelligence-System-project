# Delay Severity Analysis

This analysis classifies flights based on operational disruption severity using arrival delay duration.

```python
def classify_delay(delay):

    if delay <= 15:
        return 'On Time'

    elif delay <= 60:
        return 'Moderate Delay'

    else:
        return 'Severe Delay'
```

```python
import pandas as pd

active_flights = pd.read_csv('../data/processed/active_flights.csv')
active_flights['DELAY_CATEGORY'] = (
    active_flights['ARR_DELAY']
    .apply(classify_delay)
)

active_flights
```

```python
delay_distribution = (
    active_flights['DELAY_CATEGORY']
    .value_counts()
)

delay_distribution
```

```python
import matplotlib.pyplot as plt
delay_distribution.plot(kind='bar')

plt.title('Flight Delay Severity Distribution')

plt.show()
```

## Key Insights

- Most flights fall within acceptable operational thresholds.
- Severe delays represent major disruption events that may propagate across the airline network.
- Moderate delays indicate recurring operational inefficiencies affecting schedule reliability.
