## KPI NOTEBOOK FOR BUSINESS ANALYSIS

```python
import pandas as pd
import numpy as np

flights = pd.read_csv('../data/cleaned/flights_cleaned.csv')

total_flights = len(flights)

print(total_flights)
```

```python
cancellation_rate = (flights['CANCELLED'].mean()*100)
print(cancellation_rate)
```

```python
active_flights = pd.read_csv('../data/processed/active_flights.csv')

avg_arr_delay = active_flights['ARR_DELAY'].mean()

print(avg_arr_delay)
```

```python
avv_dep_delay = active_flights['DEP_DELAY'].mean()

print(avv_dep_delay)
```

```python
active_flights.groupby('AIRLINE')['ARR_DELAY'] \
.mean() \
.sort_values(ascending=False)
```

## AIRLINE PERFORMANCE ANALYSIS

```python
airline_delay = (active_flights.groupby('AIRLINE')['ARR_DELAY']
.mean()
.sort_values(ascending=False))
airline_delay.head(10)
```

Airlines with higher average arrival delays may indicate:
- operational inefficiencies,
- aircraft turnaround issues,
- or higher exposure to congested routes.

## AIRPORT CONGESTION ANALYSIS

```python
busiest_airports = (flights['ORIGIN']
.value_counts()
.head(10))

busiest_airports
```

```python
airport_delays = (active_flights.groupby('ORIGIN')['ARR_DELAY']
.mean()
.sort_values(ascending=False))

airport_delays.head(10)
```

Highly congested airports often experience:
- increased taxi times,
- departure queue delays,
- and cascading operational disruptions.

## DELAY CAUSE INTELLIGENCE

```python
delay_causes = {
    'Carrier_delay':
       active_flights['DELAY_DUE_CARRIER'].sum(),
    'Weather_delay':
       active_flights['DELAY_DUE_WEATHER'].sum(),
    'NAS_delay':
       active_flights['DELAY_DUE_NAS'].sum(),
    'Security_delay':
       active_flights['DELAY_DUE_SECURITY'].sum(),
    'Late_aircraft_delay':
       active_flights['DELAY_DUE_LATE_AIRCRAFT'].sum()
}
delay_causes
```
